```python
import pandas as pd

import plotly.express as px
import plotly.graph_objects as go
```

```python
import plotly.io as pio
import plotly.colors as colors
pio.templates.default = "plotly_white"
```

```python
data = pd.read_csv("Superstore.csv", encoding = 'latin-1')
data.head()
```

```python
data.describe()
```

```python
data.info()
```

```python
data['Order Date'] = pd.to_datetime(data['Order Date'])
data['Ship Date'] = pd.to_datetime(data['Ship Date'])
data.info()
```

```python
data['Order Month'] = data['Order Date'].dt.month
data['Order Year'] = data['Order Date'].dt.year
data['Order Day of Week'] = data['Order Date'].dt.dayofweek
data.head()
```

```python
#Quetion 1
sales_by_month = data.groupby('Order Month')['Sales'].sum().reset_index()
sales_by_month
```

```python
fig = px.line(sales_by_month, x='Order Month', y='Sales', title='Monthly Sale Analysis')
fig.show()
```

```python
#Quetion 2
#Sales by Category
sales_by_category = data.groupby('Category')['Sales'].sum().reset_index()
sales_by_category
```

```python
fig = px.pie(sales_by_category, values='Sales', names='Category', hole = 0.5, color_discrete_sequence = px.colors.qualitative.Pastel)
fig.update_traces(textposition='inside', textinfo='percent+label')
fig.update_layout(title_text='Sales Analysis by Category', title_font = dict(size=24))
fig.show()
```

```python
#Quetion 3
#Sales Analysis by Sub-Category
data.head()
```

```python
sales_by_subcategory = data.groupby('Sub-Category')['Sales'].sum().reset_index()
sales_by_subcategory
```

```python
fig = px.bar(sales_by_subcategory, x='Sub-Category', y='Sales', title='Sales Analysis by Sub-Category')
fig.show()
```

```python
#Quetion 4
#Monthly Profit Analysis
profit_by_month = data.groupby('Order Month')['Profit'].sum().reset_index()
profit_by_month
```

```python
fig = px.line(profit_by_month, x='Order Month', y='Profit', title='Monthly Profit Analysis')
fig.show()
```

```python
#Quetion 5
#Profit by Category
profit_by_category = data.groupby('Category')['Profit'].sum().reset_index()
profit_by_category
```

```python
fig = px.pie(profit_by_category, values='Profit', names='Category', hole = 0.5, color_discrete_sequence = px.colors.qualitative.Pastel)

fig.update_traces(textposition='inside', textinfo='percent+label')
fig.update_layout(title_text='Profit Analysis by Category', title_font = dict(size=24))
fig.show()
```

```python
#Quetion 6
#Profit by Sub-category
profit_by_subcategory = data.groupby('Sub-Category')['Profit'].sum().reset_index()
profit_by_subcategory
```

```python
fig = px.bar(profit_by_subcategory, x='Sub-Category', y='Profit', title='Profit Analysis by Sub-Category', color='Profit')
fig.show()
```

```python
#Quetion 7
#Sales & Profit - Customer Segment
sales_profit_by_segment = data.groupby('Segment').agg({'Sales': 'sum', 'Profit': 'sum'}).reset_index()
sales_profit_by_segment
```

```python
color_palette = colors.qualitative.Pastel
fig = go.Figure()
fig.add_trace(go.Bar(x=sales_profit_by_segment['Segment'], y=sales_profit_by_segment['Sales'], name='Sales', marker_color=color_palette[5]))
fig.add_trace(go.Bar(x=sales_profit_by_segment['Segment'], y=sales_profit_by_segment['Profit'], name='Profit', marker_color=color_palette[10]))
fig.update_layout(title_text='Sales & Profit - Customer Segment', xaxis_title='Sales & Profit Analysis - Customer Segment', yaxis_title='Amount' , title_font = dict(size=24))
fig.show()
```

```python
#Quetion 8
#Sales to Profit Ratio
sales_profit_segment = data.groupby('Segment').agg({'Sales': 'sum', 'Profit': 'sum'}).reset_index()
sales_profit_segment['Sales_to_Profit_Ratio'] = sales_profit_segment['Sales'] / sales_profit_segment['Profit']
print(sales_profit_segment[['Segment','Sales_to_Profit_Ratio']])
```

```python
fig = px.bar(sales_profit_segment,
             x='Segment',
             y='Sales_to_Profit_Ratio',
             title='Sales to Profit Ratio by Segment',
             labels={'Sales_to_Profit_Ratio': 'Sales to Profit Ratio', 'Segment': 'Segment'},
             color='Sales_to_Profit_Ratio',
             color_continuous_scale='Viridis')

fig.show()
```

```python
#Quetion 9
#Sales by Region
sales_by_region = data.groupby('Region')['Sales'].sum().reset_index()
sales_by_region
```

```python
fig = px.pie(sales_by_region, values='Sales', names='Region', hole = 0.5, color_discrete_sequence = px.colors.qualitative.Pastel)
fig.update_traces(textposition='inside', textinfo='percent+label')
fig.update_layout(title_text='Sales Analysis by Region', title_font = dict(size=24))
fig.show()
```

```python
#Quetion 10
# Grouping by Discount Range and summing Sales and Profit
bins = [0, 0.1, 0.5, 1]
labels = ['0%', '0.1-0.5%', '0.6-1%']
data['Discount Range'] = pd.cut(data['Discount'], bins=bins, labels=labels, right=False)

discount_impact = data.groupby('Discount Range')[['Sales', 'Profit']].sum().reset_index()
discount_impact
```

```python
# Grouping by Discount Range and summing Sales and Profit
discount_impact = data.groupby('Discount Range')[['Sales', 'Profit']].sum().reset_index()

# Plotting the impact of discount on sales and profit
fig = px.bar(discount_impact, x='Discount Range', y=['Sales', 'Profit'],
             title='Sales and Profit by Discount Range',
             labels={'Sales': 'Total Sales', 'Profit': 'Total Profit'},
             barmode='group')

fig.show()
```
